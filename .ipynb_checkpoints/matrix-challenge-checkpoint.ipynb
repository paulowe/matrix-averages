```python
#Create a 10 x 10 matrix of random numbers
import numpy as np
import pandas as pd
matrix = np.random.rand(10,10)
matrix.shape
```

```python
# Create dataframe
df = pd.DataFrame(matrix)
df.describe() #to describe column-wise means
```

```python
df.apply(pd.DataFrame.describe, axis=1) #describe row-wise means
```
